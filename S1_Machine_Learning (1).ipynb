#Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data: predicting house price based on size
data = {'Size': [500, 750, 1000, 1250, 1500],
        'Price': [150000, 200000, 250000, 300000, 350000]}
df = pd.DataFrame(data)
```

```python
# Define the feature (independent variable) and target (dependent variable)
X = df[['Size']]
y = df['Price']
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Make predictions
predictions = model.predict(X)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predictions, color='red', label='Linear Regression Line')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('Simple Linear Regression: House Price Prediction')
plt.legend()
plt.show()
```

# Multiple Linear Regression

```python
# Sample data: predicting house price based on size, bedrooms, and age
data = {'Size': [500, 750, 1000, 1250, 1500],
        'Bedrooms': [1, 2, 3, 3, 4],
        'Age': [5, 10, 15, 20, 25],
        'Price': [150000, 200000, 250000, 300000, 350000]}
df = pd.DataFrame(data)
```

```python
# Define the features (independent variables) and target (dependent variable)
X = df[['Size', 'Bedrooms', 'Age']]
y = df['Price']
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Make predictions
predictions = model.predict(X)
```

```python
# Display predictions
print(predictions)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Sample data: predicting disease progression based on time
data = {'Time': [1, 2, 3, 4, 5],
        'Progression': [1, 4, 9, 16, 25]}
df = pd.DataFrame(data)
```

```python
# Define the feature (independent variable) and target (dependent variable)
X = df[['Time']]
y = df['Progression']
```

```python
# Transform features to polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# Make predictions
predictions = model.predict(X_poly)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predictions, color='red', label='Polynomial Regression Line')
plt.xlabel('Time')
plt.ylabel('Progression')
plt.title('Polynomial Regression: Disease Progression Prediction')
plt.legend()
plt.show()
```
